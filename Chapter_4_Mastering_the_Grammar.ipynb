## 4.2 Building a Scatterplot

엔진 크기(engine size)와 연비(fuel economy)는 어떤 관련이 있을까?    
실린더(cylinders)수에 따른 배기량(engine displacement)와 고속도로 연비(highway mpg)의 산점도를 통해 알아보자.

```r
library(ggplot2)
```

```r
str(mpg)
```

```r
# 실런더 수(cyl)를 점의 색으로 표현함으로써 3차원 데이터 plot을 작성
ggplot(mpg, aes(displ, hwy, colour = factor(cyl))) +
geom_point()
```

이처럼 plot은 쉽게 그릴 수 있지만 그 내부의 코드가 작동하는 원리가 무엇이고, ggplot2는 어떻게 이런 plot을 작성할 수 있는지에 대해 생각해보자.

### 4.2.1 Mapping Aesthetics to Data

산점도(scatterplot)는 각 관측치들을 point(점)으로서 나타내며 두 변수의 값을 이용하여 위치를 지정한다.

각 point에는 수평 및 수직 위치뿐만 아니라 size(크기), colour(색상) 및 shape(모양)이 있다.    
이러한 특성을 **aesthetic(미학)**이라고 부르며 이는 그래픽이 인식할 수 있는 속성(property)이다.

각 aesthetic은 변수에 매핑되거나 상수값으로 설정 될 수 있다.    
예를 들어, 위 예제 plot에서 `displ`는 horizontal position(수평 위치)로, `hwy`는 vertical position(수직 위치)로 `cyl`는 colour(색상)으로 매핑된 것을 볼 수 있다.    
여기서 Size(크기)와 shape(모양)은 변수에 매핑되지는 않고 default values(기본값 : 상수)로 유지된다.

이러한 매핑이 있으면 아래와 같은 aesthetic 정보를 기록하는 새로운 데이터 세트를 만들 수 있다.

**mapped aesthetic data**

|$x$(horizontal)|$y$(vertical)|colour|
|:--:|:--:|:--:|
|1.8|29|4|
|1.8|29|4|
|2.0|31|4|
|2.0|30|4|
|2.8|26|6|
|2.8|26|6|
|3.1|26|6|
|1.8|26|4|

이 데이터 세트는 원래 데이터에 aesthetic 매핑을 적용한 결과이다.

이 데이터를 사용하여 다양한 유형의 plot을 만들 수 있다.     
예를 들어, scatter plot는 point을 사용하지만 점 대신에 선을 그려 line plot을 만들 수 있다.(bar를 사용한다면 bar plot이 된다.) 이러한 plot들은 데이터에 대해서 의미가 없지만 그릴 수 있다는 것을 보여준다.

```r
# line plot
ggplot(mpg, aes(displ, hwy, colour = factor(cyl))) +
geom_line()
```

```r
# bar plot
ggplot(mpg, aes(displ, hwy, colour = factor(cyl))) +
geom_bar(stat = "identity", position = "identity", fill = NA)
```

이런 방법으로 ggplot에서 의미는 없지만 문법적으로 유효한 많은 plot들을 작성 할 수 있다.

---

여기서 등장한 point(점), line(선) 및 bar(막대)는 모두 **기하학적 객체(geom, 기하구조)**이다.    
geom은 plot의 'type(유형)'을 결정한다. 특정한 하나의 geom을 사용하는 plot에는 종종 특별한 이름이 있다.

|Named plot|Geom|Other features|
|:--:|:--:|:--:|
|Scatter plot|Point||
|Bubble chart|Point|Size mapped to a variable|
|Bar chart|Bar||
|Box-and-whisker plot|Box plot||
|Line chart|Line||

----

여러 geom의 조합으로 구성된 복잡한 plot에는 특별한 이름이 없으므로 직접 작성해야 한다.    
예를 들어, 아래 plot은 scatter plot 위에 그룹 회귀선을 overlay 한 것이다.

```r
ggplot(mpg, aes(displ, hwy, colour = factor(cyl))) +
geom_point() +
geom_smooth(method = "lm")
```

이 plot을 부르는 특별한 이름은 없다.

문법을 숙지하고 나면 우리는 특정한 문제에 맞게 plot을 작성하게 되므로 특별한 이름이 필요가 없을 것이다.

### 4.2.2 Scaling

우리의 데이터 값은 컴퓨터에는 아무런 의미를 가지지 않는다. 따라서, 데이터 단위(리터, 마일 당 갤런과 실린더 수 등)를 컴퓨터가 display할 수 있는 그래픽 단위(픽셀과 색상 등)으로 변환해야 한다.

이러한 과정을 **scaling**이라고 부르며 scale에 의해 수행된다.
> 이 예제에서는 3가지 aesthetic이 scale되야 한다:    
- horizontal position(x)
- vertical position(y)
- colour

**horizontal position(x), vertical position(y) Scaling**

여기서는 default linear(기본 선형) scale을 사용하기 때문에 Scaling position을 쉽게 지정할 수 있다.    
따라서 x와 y에 대해 $[0, 1]$로의 linear mapping만 필요하다.

>ggplot2의 drawing system(**grid**)이 최종 변환을 처리하기 때문에 정확한 pixel 대신에 $[0, 1]$을 사용한다.

다음으로 coordinate system(좌표계, **coord**)를 통해 두 위치(x, y)를 결합하여 plot의 최종 위치를 결정한다.

>대부분의 경우 Cartesian coordinates(데카르트 좌표계, 직교 좌표계)이지만     
polar coordinate(극 좌표) 또는 지도에 사용되는 spherical projection(구형 투영)일 수 있다.


**colour Scaling**

colour을 mapping한 결과는 숫자가 아니기 때문에(colour, 색상값) 더 복잡하지만 colour는 사람의 눈이 색상을 감지하는 3가지 세포와 대응되는 3가지 요소를 가진다고 생각할 수 있으므로 이러한 3가지 세포 유형은 3차원의 colour 공간을 만든다. 

*colour Scaling은 데이터 값을 colour 공간의 점에 매핑하는 작업을 포함한다.*

> 이를 수행하는 방법은 여러 가지가 있지만 여기서는 `cyl`가 범주형(categorical) 변수이기 때문에 colour wheel(색상환)에 균등하게 간격을 둔 hue(색상)으로 값을 매핑한다.

## 4.3 Adding Complexity

앞서 이용한 간단한 예제를 통해 좀 더 복잡한 plot을 그려보자.

```r
ggplot(mpg, aes(displ, hwy)) +
geom_point() +
geom_smooth() +
facet_wrap(~year)
```

이 plot은 세 가지 새로운 구성 요소를 혼합해서 추가한 것이다:
- facet(패싯, 측면)
- multiple layer(다중 레이어)
- statistic(통계)

> `facet`과 `layer`는 위에서 설명한 데이터 구조를 확장한다: 각 `layer`의 각 `facet panel`에는 자체 데이터 집합이 있다.(이는 3차원 배열로 생각할 수 있다.) 
- `facet`의 panel은 2차원 grid를 형성하고, `layer`는 3번째 차원으로 2차원 grid 위에 확장된다.    
*이 경우에는 `layer`의 데이터가 동일하지만 일반적으로는 서로 다른 `layer`에 다른 데이터 세트로 plot할 수 있다.

---

`smooth layer`는 raw data를 표시한 것이 아니라 데이터의 'statistical transformation(통계적 변환)'을 표시한 것이기 때문에 `point layer`와 다르다.    
*여기서 `smooth layer`는 데이터의 중앙을 통과하는 'smooth line(평활선)'을 나타낸다

이렇게하려면 앞서 설명된 과정에 추가적인 단계가 필요하다.
- 데이터를 aesthetic으로 mapping 한 후에, 데이터를 'statistical transformation'이나 `stat`를 통해 유용한 방식으로 데이터를 조작한다.
> 이 예에서의 `stat`는 loess smoother(평활선)으로 데이터를 fit한 다음 데이터 범위 내의 고르게 이격된 지점에서 예측을 반환한다.    
*다른 유용한 `stat`는 1 및 2 차원binning, 그룹 평균, 분위수 회귀 및 등고선이 있다.

`loess` : 변동성이 존재하는 시계열을 평활화하기 위해 일반적으로 사용하는 방법 중의 하나이다. 이 방법은 전체 데이터 중의 국소 부분에 최소자승회귀를 적용하는 비모수적 방법으로 수치 벡터들을 평활하기에 적합한 방법이다. LOESS는 local regression의 약어로서 데이터의 어떤 점을 기준으로 그 근방의 값들에 다중회귀를 실행하는 비모수적 접근이다.

이러한 추가단계 뿐만 아니라 `scale`을 하기위해서도 몇가지 추가적인 단계가 더 필요하다.

이 예제에서 여러 `facet`과 `layer`에 대해서 여러 개의 데이터 세트를 가지고 있기 때문에 이들이 동일한 `scale`을 가지도록 해야한다.

이러한 Scaling은 세 part으로 이루어진다.
1. transforming(변환)
2. training(훈련)
3. mapping

#### 1. Scale transformation 

`Scale transformation`은 'statistical transformation' 전에 수행하므로 scale transformed data에서 통계 계산을 수행한다.

*이렇게하면 linear scales에서 log(x) vs. log (y)의 plot이 log scale의 x와 y가 동일하게 보인다.

square roots, logarithms and reciprocals 등 사용할 수있는 많은 변형이 있다.(6.6.1에 더 자세히 다룬다.)

**2. training**

통계 계산이 된 후 각 scale은 모든 layer 및 facets의 모든 데이터 집합에 대해 학습된다.

training 연산은 전체 데이터의 범위를 가져오기 위해 개별 데이터 세트의 범위를 결합하는 것이다.

> 이 단계가 없으면 `scale`은 부분적으로만 이해할 수 있으며 위치가 맞지 않기때문에 다른 layer를 overlay 할 수 없다.

**3. mapping**

마지막으로 `scale`은 데이터 값을 aesthetic 값으로 mapping 한다. 이 작업은 local operation 이다.
> 각 데이터 세트의 변수는 aesthetic 값에 매핑되어 geom에 의해 렌더링 될 수있는 새로운 데이터 세트를 생성한다.

**전체 과정**

각 사각형은 layer를 나타내며,이 schematic(도식)은 3개의 layer와 3개의 panel이 있는 plot을 나타낸다.

## 4.4 Components of the Layered Grammar

앞에서 plot을 구성하는 몇 가지 component들을 살펴 보았다.

- `data`와 `aesthetic mappings`
- `geometric objects(geoms)` : 기하학적 객체(기하구조)
- `statistical transformations(stats)` : 통계적 변환
- `scale`
- `facetting` : 측면보기

*추가적으로 coordinate system(좌표계)를 다루었다.

또한 여기서는 다루지 않은 graphic objects(그래픽 객체)를 겹치는 `position adjustment(위치 조정)`이 있다.

---

`data`, `mapping`, `stat`, `geom`, `position adjustment`들은 하나의 layer를 형성한다.

하나의 plot은 여러 개의 layer를 가질 수 있다.(scatter plot에서 smoothed line(평활화된 선)을 겹쳐 놓은 예제에서와 같이)

이들 전부에 대해 layered grammar는 다음의 조합으로 plot을 정의한다.
- default dataset과 변수에서 aesthetics으로의 mapping 집합들
- 하나 이상의 layers(`geometric object`, `statistical transformation`, `position adjustment`와 선택적으로 `dataset`, `aesthetic mappings`로 구성된)
- 각각의 aesthetic mapping을 위한 하나의 scale.
- facetting 규격

다음 절에서는 각 상위 구성 요소를 보다 정확하게 설명한다.

### 4.4.1 Layers

우리가 plot에서 인식하는 객체를 생성한다.

다섯 부분으로 구성된다.
1. Data
2. Aesthetic mappings.
3. A statistical transformation (stat).
4. A geometric object (geom).
5. A position adjustment.

layer에 대한 속성은 Chapter5에서 다룰 것이다.

### 4.4.2 Scales

데이터를 `aesthetic 속성`으로 mapping하는 것을 제어하며 plot에 사용된 모든 aesthetic을 위한 scale가 필요하다.

각 scale는 plot의 모든 데이터에서 연산하여 데이터를 aesthetics로 일관되게 mapping한다.

4가지 다른 'scale'의 범례에 대한 예제

왼쪽 : continuous variable(연속형 변수)는 size와 colour로 mapping

오른쪽 : discrete variable(이산형 변수는) shape와 colour로 mapping

![fig4_3](https://github.com/JJongKwon/ggplot2/blob/master/fig4_3.PNG?raw=true)

**scale는 parameter의 집합을 가진 함수 및 역함수이다.**

예를 들어, colour gradient(그라데이션) scale은 실제 선의 분할을 colour space을 통과하는 경로로 mapping한다.

함수의 `parameters`는 경로가 linear(선형)인지 곡선인지, 사용할 colour space(e.g., LUV 또는 RGB) 및 시작과 끝의 색상을 정의한다.

그래프에서 값을 읽을 수 있는 guide(안내선)를 그리는 데 역함수를 사용한다.
- guide는 axes(축 : position scales의 경우) 또는 legend(범례, 다른 모든 경우)

대부분의 mappings에는 고유한 역함수가 존재지만(즉, mapping 함수는 일대일 대응). 그렇지 않는 경우도 많다.

고유한 역함수를 사용하면 원본 데이터를 구할 수 있지만, 단일 측면에 집중하고자하는 경우 항상 바람직한 것은 아니다.(6장에서 자세히)

### 4.4.3 Coordinate System

`coordinate system(coord)`를 간단히 말하면 plot의 평면에 객체의 'position'를 mapping하는 것이다.

> 'Position'는 두 좌표(x, y)로 지정되는 경우가 많지만 3개 이상일 수도 있다.    
(ggplot2에서는 구현되지 않는다).

> Cartesian coordinate system(데카르트 좌표계)는 2차원에 대한 가장 일반적인 좌표계이다.    
(극좌표 및 다양한 지도 투영은 덜 사용한다.)

좌표계는 모든 position variables(위치 변수)에 동시에 영향을 주며 geometric object(기하 객체)의 모양을 변경한다는 점에서 scale과는 다르다.(예를 들어, 극 좌표계에서 bar geom은 원의 segment처럼 보인다.)

또한 scaling은 statistical transformation(통계 변환) 전에 수행하는 반면에, coordinate transformation(좌표 변환)은 통계 변환 후에 한다.(결과는 7.5절에)

좌표계는 axes와 grid line을 그리는 방법을 제어한다.

### 4.4.4 Facetting

우리가 일반적인 framework(작업틀)에 포함시켜야 할 충분한 유용성이 있는 것이 있다.: `facetting`(조건부 또는 trellised plot의 일반적인 경우)

이를 통해 전체 데이터 집합의 다른 하위 집합을 개별적으로 볼 수있다.

`facetting`은 어떤 특정한 패턴이 데이터의 모든 경우에 적용되는지 여부를 조사 할 때 강력한 도구이다.

`facetting` specification은 데이터를 분할하는 데 사용되는 변수와 position scales이 자유롭거나 제한되어야하는지 여부를 설명한다.

`facetting`에 대해서는 7장에서 다룬다.
